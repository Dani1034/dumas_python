```python
for line in lines[9000:9020]:
    print(line, end="")
```

```python
comte_text1 = ''

for line in lines:
    comte_text1 += line

print(len(comte_text1))
```

```python
import re
 
def split_into_chapters(text):
    # Regular expression to match chapter headings
    pattern = r"\n{2,}([IVXLCDM]+)\n+\s*(.*?)\n"
   
    # Split text while keeping the chapter titles
    matches = re.split(pattern, text)
 
    # The first item (matches[0]) will be the text before the first chapter, so we ignore it
    chapters = {}
    for i in range(1, len(matches), 3):
        chapter_num = matches[i]
        chapter_title = matches[i + 1]
        chapter_content = matches[i + 2].strip()
        chapters[f"{chapter_num}"] = chapter_content
 
    return chapters
 
chapters = split_into_chapters(comte_text1)
 
# Print chapter titles
for title, content in chapters.items():
    print(title)
    print(content[:500])  # Print the first 500 characters of each chapter
    print("-" * 50)
```

```python
import spacy
nlp = spacy.load('fr_core_news_sm')
```

```python
doc1 = nlp(comte_text1)
```

```python
def show_ents(doc1):
    if doc1.ents:
        for ent in doc1.ents:
            print(ent.text + ' - ' + ent.label_ + ' - ' + str(spacy.explain(ent.label_)))
```

```python
show_ents(doc1)
```

```python
def calculate_freqs(chapters, target_ent):
    freqs = {}
    for number, chapter in chapters.items():
        doc = nlp(chapter)
        freqs[number] = 0
        for ent in doc.ents:
            if ent.text == target_ent:
                #print(ent.text)
                freqs[number] += 1
    return freqs
```

```python
freqs_1 = calculate_freqs(chapters_1, 'Mercédès')
freqs_2 = calculate_freqs(chapters_2, 'Mercédès')
freqs_3 = calculate_freqs(chapters_3, 'Mercédès')
freqs_4 = calculate_freqs(chapters_4, 'Mercédès')
```

```python
all_Mfreqs = {**freqs_1, **freqs_2, **freqs_3, **freqs_4}

print(all_Mfreqs)
```
